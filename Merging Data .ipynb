```python
import numpy as np
import pandas as pd
from pandas import  DataFrame,Series
```

```python
dframe1=DataFrame({"Key":["X","Y","Z","Y","W","Z"],"Values":np.arange(6)})
```

```python
dframe1
```

```python
dframe2=DataFrame({"Key":["A","Y","Z"],"Values":[1,2,3]})
dframe2
```

```python
pd.merge(dframe1,dframe2,on="Key")
```

```python
pd.merge(dframe1,dframe2,on="Key",how="right")
```

```python
pd.merge(dframe1,dframe2,on="Key",how="outer")
```

```python
rightdf=DataFrame({"Key1":["NY","NY","LA","LA"],
                   "Key2":["one","two","three","four"],
                   "Values1":[10,20,30,40]})
```

```python
Leftdf=DataFrame({"Key1":["La","La","CHI","CHI","NY"],
                   "Key2":["one","two","three","four","five"],
                   "Values1":[50,60,70,80,90]})
```

```python
Leftdf
```

```python
pd.merge(rightdf,Leftdf,on=["Key1","Key2"],how="outer",suffixes=["rigthy","lefty"])
```

```python
keydf=DataFrame({"key1":["X","Y","Z"],
                 "Data": range(3)  })

```

```python
indexdf= DataFrame({"goup_data":[10,20]},index=["X","Y"])
```

```python
pd.merge(keydf,indexdf,left_on="key1",right_index=True)
```

```python
keydf.join(indexdf,how="outer")
```

```python
df_left_hr = DataFrame({'key1': ['SF','SF','SF','LA','LA'],
                   'key2': [10, 20, 30, 20, 30],
                   'data_set': np.arange(5.)})
df_right_hr = DataFrame(np.arange(10).reshape((5, 2)),
                   index=[['LA','LA','SF','SF','SF'],
                          [20, 10, 10, 10, 20]],
                   columns=['col_1', 'col_2'])
```

```python
pd.merge(df_left_hr,df_right_hr,left_on=["key1","key2"],right_index=True)
```
