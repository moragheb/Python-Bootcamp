```python
class Match:
    def __init__(self,player1="",player2="",startTime="",startdate="",player1score="0",player2score="0",matchstatus="",sport="",lround="",methodofvictory="",matchpar="72",matchrank="0",player1outs="0",player1overs="0",player2outs="0",player2overs="0"):
        self.player1=player1
        self.player2=player2
        self.startTime=startTime
        self.player1score=player1score
        self.player2score=player2score
        self.matchstatus=matchstatus
        self.sport=sport
        self.startdate=startdate
        self.lround=lround
        self.methodofvictory=methodofvictory
        self.matchpar=matchpar
        self.matchrank=matchrank
        self.player1outs=player1outs
        self.player1overs=player1overs
        self.player2outs=player2outs
        self.player2overs=player2overs
    def __str__(self):
        if self.sport== "cricket" and self.matchstatus!="Not Started" :
            return self.player1+ " :score :" + self.player1score + " and outs : "+self.player1outs +" and over is  :" + self.player1overs + " "+ self.player2 +" :score:  " + self.player2score + " outs  :" + self.player2outs + " overs : " +self.player2overs 
        if self.sport == "mma" or self.sport =="boxing":
            if(self.matchstatus!="Not Started"):
                if self.player1score=="winner":
                  return "player "+ self.player1 + " beats player " +self.player2 +" by "+ self.methodofvictory
                else:
                  return "player "+ self.player2 + " beats player " +self.player1 +" by "+ self.methodofvictory
            else:
              return str(self.lround) +" "+ self.startdate + self.startTime + " Team1:  " + self.player1 +" Team 2: " + self.player2
        else :
            if(self.matchstatus!="Not Started"):
               return str(self.lround) +" current status:" + self.matchstatus + " Home Team is:  " + self.player1+" score is:"+ str(self.player1score) +"   and Away Team: " + self.player2  + "  score is " + str(self.player2score)
            else:
               return str(self.lround) +" "+ self.startdate + " "+self.startTime + " Team1:  " + self.player1 +" Team 2: " + self.player2
        
```

```python
def get_alleaguefixturesorresults(sport,country,leaguename,season="2019",fixtures=True):
     #driver = webdriver.PhantomJS()
        options = webdriver.ChromeOptions() 
        options.add_argument('profile-directory=Default')
        options.add_argument("--headless")
        options.add_argument("user-data-dir=C:\\Users\\Mo Ragheb\\AppData\\Local\\Google\\Chrome\\User Data\\Profile 1") #Path to your chrome profile
        driver=webdriver.Chrome("C:\\Program Files (x86)\\Google\\Chrome\\chromedriver_win32\\chromedriver.exe",options=options)
        requesturl=""
        logging.info(" start exporting all games for  sport :{0}  country :{1}  league name:{2} season:{3}".format(sport,country,leaguename,season))
        if sport=="hockey" or   leaguename =="ipl" :
            requesturl=flashurl+leaguename
        else:
            requesturl=flashurl+sport+"/"+country+"/"+leaguename
            
            
        if fixtures:
            driver.get(requesturl+"/fixtures/")
        else:
            driver.get(requesturl+"/results/")
        while True:
            try:
                
                driver.find_element_by_class_name('event__more--static').click()
                sleep(5)
            except Exception as e:
                 logging.error("Error ocured :{0}".format(str(e)))   
                 break
        logging.info("url openned is :{0}".format(driver.current_url))
        
        soup=BeautifulSoup(driver.page_source,"lxml")
        comp=Comp(competitionname=leaguename,sport=sport,season=season,country=country)
        
        match=Match()
        currentround=1
        nomatchtag=soup.find("div",class_="no-match-found nmf")
        if nomatchtag!=None:
            logging.info("no matches schudeld yet for this league")
            return None

        
        tag= soup.find("div",class_="sportName "+sport)
        
        for child in tag.findChildren():
             if child.name=="div" and  "event__round" in child.attrs["class"]:
                    currentround=child.text
             if child.name=="div" and "event__time"  in child.attrs["class"]:
                    if fixtures:
                        match.matchstatus="Not Started"
                    else :
                        match.matchstatus="Finished"
                    tempdate=child.text.replace("FRO","").rsplit(".",1)[0]
                    temptime=child.text.replace("FRO","").rsplit(".",1)[1]
                    match.startTime=tempdate+" "+temptime
                    for sibling in child.next_siblings:
                        if "event__participant--home" in sibling.attrs["class"]:
                            match.player1=sibling.text
                        if "event__participant--away" in sibling.attrs["class"]:
                            match.player2=sibling.text
                        
                        if sibling.name=="div" and "event__score--away" in sibling.attrs["class"]:
                            match.player2score=sibling.text
                        if sibling.name=="div" and "event__score--home" in sibling.attrs["class"]:
                            match.player1score=sibling.text
                        if sibling.name=="div" and "event__scores" in sibling.attrs["class"]  and not fixtures:
                            tmp_scores=sibling.text.split("-")
                            match.player1score=tmp_scores[0]
                            match.player2score=tmp_scores[1]
                           
                           
                           
                            

                    match.sport=sport
                    match.lround=currentround
                    comp.matches.append(match)
                    logging.info("match added {0}".format(str(match)))
                    match=Match()
                    pass
                        
        return comp
        driver.quit()
```

```python

def replaceMultiple(mainString, toBeReplaces, newString):
    # Iterate over the strings to be replaced
    for elem in toBeReplaces :
        # Check if string is in the main string
        if elem in mainString :
            # Replace the string
            mainString = mainString.replace(elem, newString)
    mainString=mainString.replace("--","-")
    return  mainString
```

```python
def as_text(value):
    if value is None:
        return ""
    return str(value)
```

```python
def adjustwidth(worksheet):
    for column_cells in worksheet.columns:
        length = max(len(as_text(cell.value)) for cell in column_cells)
        worksheet.column_dimensions[column_cells[0].column].width = length+3
```

```python
def sheet_contains(match,sheet):
    matchtime=parser.parse(match.startdate+" "+match.startTime,dayfirst=True).date()

    for row in sheet :
        try:
            sheettime=parser.parse(row[4].value,dayfirst=True ).date()
        except:
            sheettime=row[4].value                             
        if match.player1 == row[1].value and match.player2==row[3].value and (matchtime==sheettime or (matchtime == (sheettime + datetime.timedelta(days=1)))):
            return True 

    return False
        
```

```python
def row_needsupdating(match,row):
    matchdate=parser.parse(match.startdate+" "+match.startTime,dayfirst=True).date()
    try:
            sheetdate=parser.parse(row[4].value ,dayfirst=True).date()
    except:
            sheetdate=row[4].value                             
    if match.player1 == row[1].value and match.player2==row[3].value and matchdate==sheetdate and row[12].value!="Updated":
            return True 
    elif match.player1 == row[1].value and match.player2==row[3].value and ((matchdate==(sheetdate + datetime.timedelta(days=1))) or (matchdate==(sheetdate -datetime.timedelta(days=1)))) and row[12].value!="Updated":
            return True
    return False
```

```python
def checkwinner(sport,player1result,player2result):

    if sport=="soccer" or sport=="basketball" or sport== "tennis" or sport== "baseball" or sport=="cricket" or sport=="american-football":
        if int(player1result)> int(player2result):
            return 1
        elif int(player1result)<int(player2result):
            return 2
        else :
            return 0
    elif sport=="mma" or sport=="boxing":
        if player1result=="winner":
              return 1
        else:
              return 2
```
